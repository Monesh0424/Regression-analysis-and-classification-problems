```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
housing = pd.read_csv(r'C:\Users\USER\Desktop\python\ML\2. Multiple Linear Regression\Multiple_Linear_Regression\Housing.csv')
housing.head()
```

```python
housing['mainroad'] = housing['mainroad'].map({'yes':1 , 'no':0})
housing['guestroom'] = housing['guestroom'].map({'yes':1 , 'no':0})
housing['basement'] = housing['basement'].map({'yes':1 , 'no':0})
housing['hotwaterheating'] = housing['hotwaterheating'].map({'yes':1 , 'no':0})
housing['airconditioning'] = housing['airconditioning'].map({'yes':1 , 'no':0})
housing['prefarea'] = housing['prefarea'].map({'yes':1 , 'no':0})
```

```python
housing.head()
```

```python
df = pd.get_dummies(housing['furnishingstatus'] , drop_first=True)
```

```python
housing = pd.concat([df,housing],axis=1)
```

```python
housing.drop('furnishingstatus',inplace=True,axis=1)
```

```python
housing.head()
```

```python
# Is bathroom available for every bedroom ?

housing['bbratio'] = housing['bedrooms'] / housing['bathrooms']
```

```python
housing.head()
```

```python
# Deriving area per bedroom based on area and bedroom

housing['area_per_bedroom'] = housing['area']/housing['bedrooms']
```

```python
housing.head()
```

```python
# Scaling

def norm(x):
    return (x-min(x))/(max(x)-min(x))
```

```python
housing = housing.apply(norm)
```

```python
housing.head()
```

```python
X_features = list(housing.columns)
X_features.remove('price')
```

```python
import statsmodels.api as sm
```

```python
X = sm.add_constant(housing[X_features])
y = housing['price']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=100)
```

```python
# Model

lm = sm.OLS(y_train,X_train).fit()
```

```python
# Accuracy

lm.summary2()
```

```python
# Test for Overfit

# RMSE on training set

from sklearn import metrics

ytrain_pred = lm.predict(X_train)
rmse_train = np.sqrt(metrics.mean_squared_error(y_train,ytrain_pred))
print('rmse for training data is {}'.format(rmse_train))


# RMSE in testing set

ytest_pred = lm.predict(X_test)
rmse_test = np.sqrt(metrics.mean_squared_error(y_test,ytest_pred))
print('rmse for testing data is {}'.format(rmse_test))
```

# LASSO REGRESSION

```python
from sklearn.linear_model import Lasso
```

```python
# Applying alpha = 1 and running algorithms for 500 times  
lasso = Lasso(alpha=0.01 , max_iter=500)
```

```python
lasso = lasso.fit(X_train,y_train)
lasso
```

```python
# Test for Overfit after applying L1 Norm

# RMSE on training set

from sklearn import metrics

lasso_ytrain_pred = lasso.predict(X_train)
lasso_rmse_train = np.sqrt(metrics.mean_squared_error(y_train,lasso_ytrain_pred))
print('rmse for training data after applying lasso is {}'.format(lasso_rmse_train))


# RMSE in testing set

lasso_ytest_pred = lasso.predict(X_test)
lasso_rmse_test = np.sqrt(metrics.mean_squared_error(y_test,lasso_ytest_pred))
print('rmse for testing data after applying lasso is {}'.format(lasso_rmse_test))
```

```python
lasso_coef = pd.DataFrame({'columns':housing.columns , 'coef':lasso.coef_})
```
